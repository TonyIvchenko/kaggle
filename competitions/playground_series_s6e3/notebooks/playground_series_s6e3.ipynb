# Playground Series S6E3

Quick notebook for schema checks, holdout metrics, and submission preview.

```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path('..').resolve()
RAW = ROOT / 'data' / 'raw'
MODELS = ROOT / 'models'
SUBMISSIONS = ROOT / 'submissions'
```

```python
sorted(path.name for path in RAW.glob('*.csv'))
```

```python
train = pd.read_csv(RAW / 'train.csv')
test = pd.read_csv(RAW / 'test.csv')
sample = pd.read_csv(RAW / 'sample_submission.csv')

print('train shape:', train.shape)
print('test shape:', test.shape)
print('sample shape:', sample.shape)
display(train.head())
display(test.head())
display(sample.head())
```

```python
metrics_path = MODELS / 'playground_series_s6e3_metrics.json'
if metrics_path.exists():
    metrics = json.loads(metrics_path.read_text())
    print('task_type:', metrics.get('task_type'))
    print('selected strategy:', metrics.get('selection', {}).get('selected_strategy'))
    display(pd.DataFrame(metrics.get('holdout_metrics', {}).get('strategy_metrics', [])))
else:
    print('Run scripts/train_model.py to generate metrics first.')
```

```python
submission_path = SUBMISSIONS / 'submission.csv'
if submission_path.exists():
    display(pd.read_csv(submission_path).head(20))
else:
    print('Run scripts/train_model.py to generate submission first.')
```
