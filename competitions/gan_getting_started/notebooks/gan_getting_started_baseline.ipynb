# GAN Getting Started Baseline

Creates a valid `images.zip` submission by copying `photo_jpg` images and renaming to `00000.jpg`, `00001.jpg`, ...

This is only a submission-format baseline, not a competitive GAN.

```python
from pathlib import Path
from zipfile import ZIP_DEFLATED, ZipFile

SOURCE_DIR = Path('/kaggle/input/gan-getting-started/photo_jpg')
OUTPUT_ZIP = Path('/kaggle/working/images.zip')
NUM_IMAGES = 7000

images = sorted(SOURCE_DIR.glob('*.jpg'))
if not images:
    raise RuntimeError(f'No jpg files found under {SOURCE_DIR}')

if OUTPUT_ZIP.exists():
    OUTPUT_ZIP.unlink()

with ZipFile(OUTPUT_ZIP, mode='w', compression=ZIP_DEFLATED) as archive:
    for idx in range(NUM_IMAGES):
        src = images[idx % len(images)]
        archive.write(src, arcname=f'{idx:05d}.jpg')

size_mb = OUTPUT_ZIP.stat().st_size / (1024 * 1024)
print(f'Created: {OUTPUT_ZIP}')
print(f'Source images: {len(images)}')
print(f'Written images: {NUM_IMAGES}')
print(f'Zip size: {size_mb:.2f} MB')
```
